Attribute Information:

1. name: Name of the country concerned
2. landmass: 1=N.America, 2=S.America, 3=Europe, 4=Africa, 4=Asia, 6=Oceania
3. zone: Geographic quadrant, based on Greenwich and the Equator; 1=NE, 2=SE, 3=SW, 4=NW
4. area: in thousands of square km
5. population: in round millions
6. language: 1=English, 2=Spanish, 3=French, 4=German, 5=Slavic, 6=Other Indo-European, 7=Chinese, 8=Arabic, 9=Japanese/Turkish/Finnish/Magyar, 10=Others
7. religion: 0=Catholic, 1=Other Christian, 2=Muslim, 3=Buddhist, 4=Hindu, 5=Ethnic, 6=Marxist, 7=Others
8. bars: Number of vertical bars in the flag
9. stripes: Number of horizontal stripes in the flag
10. colours: Number of different colours in the flag
11. red: 0 if red absent, 1 if red present in the flag
12. green: same for green
13. blue: same for blue
14. gold: same for gold (also yellow)
15. white: same for white
16. black: same for black
17. orange: same for orange (also brown)
18. mainhue: predominant colour in the flag (tie-breaks decided by taking the topmost hue, if that fails then the most central hue, and if that fails the leftmost hue)
19. circles: Number of circles in the flag
20. crosses: Number of (upright) crosses
21. saltires: Number of diagonal crosses
22. quarters: Number of quartered sections
23. sunstars: Number of sun or star symbols
24. crescent: 1 if a crescent moon symbol present, else 0
25. triangle: 1 if any triangles present, 0 otherwise
26. icon: 1 if an inanimate image present (e.g., a boat), otherwise 0
27. animate: 1 if an animate image (e.g., an eagle, a tree, a human hand) present, 0 otherwise
28. text: 1 if any letters or writing on the flag (e.g., a motto or slogan), 0 otherwise
29. topleft: colour in the top-left corner (moving right to decide tie-breaks)
30. botright: Colour in the bottom-left corner (moving left to decide tie-breaks)

```python
attributes = ['name','landmass','zone','area','population','language','religion','bars','stripes','colours','red','green',
              'blue','gold','white','black','orange','mainhue','circles','crosses','saltires','quarters','sunstars','crescent',
              'triangle','icon','animate','text','topleft','botright']
```

```python
data = pd.read_csv('flag.data', names=attributes)
data.head()
```

```python
data_core = data.drop(['name', 'mainhue', 'topleft', 'botright'], axis=1)
data_core.head()
```

```python
def religions(data):
    if data == 1:
        return 0 #Christian 
    elif data == 2: #Muslims
        return 1
    else:
        return 2 #Others
```

```python
target = data['religion'].apply(religions)
target
```

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.RobustScaler()
robust_df = scaler.fit_transform(data_core)
robust_df = pd.DataFrame(robust_df, columns=data_core.columns)
robust_df.head()
```

```python
scaler = preprocessing.StandardScaler()
standard_df = scaler.fit_transform(data_core)
standard_df = pd.DataFrame(standard_df, columns=data_core.columns)
standard_df.head()
```

```python
scaler = preprocessing.MinMaxScaler()
minmax_df = scaler.fit_transform(data_core)
minmax_df = pd.DataFrame(minmax_df, columns=data_core.columns)
minmax_df.head()
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(ncols = 4, figsize =(20, 5))

ax1.set_title('Before Scaling')
sns.kdeplot(data_core['landmass'], ax = ax1, color ='red')
sns.kdeplot(data_core['zone'], ax = ax1, color ='blue')
sns.kdeplot(data_core['area'], ax = ax1, color ='green')
sns.kdeplot(data_core['population'], ax = ax1, color ='yellow')
sns.kdeplot(data_core['language'], ax = ax1, color ='orange')

ax2.set_title('After Robust Scaling')
sns.kdeplot(robust_df['landmass'], ax = ax2, color ='red')
sns.kdeplot(robust_df['zone'], ax = ax2, color ='blue')
sns.kdeplot(robust_df['area'], ax = ax2, color ='green')
sns.kdeplot(robust_df['population'], ax = ax2, color ='yellow')
sns.kdeplot(robust_df['language'], ax = ax2, color ='orange')

ax3.set_title('After Standard Scaling')
sns.kdeplot(standard_df['landmass'], ax = ax3, color ='red')
sns.kdeplot(standard_df['zone'], ax = ax3, color ='blue')
sns.kdeplot(standard_df['area'], ax = ax3, color ='green')
sns.kdeplot(standard_df['population'], ax = ax3, color ='yellow')
sns.kdeplot(standard_df['language'], ax = ax3, color ='orange')

ax4.set_title('After Min-Max Scaling')
sns.kdeplot(minmax_df['landmass'], ax = ax4, color ='red')
sns.kdeplot(minmax_df['zone'], ax = ax4, color ='blue')
sns.kdeplot(minmax_df['area'], ax = ax4, color ='green')
sns.kdeplot(minmax_df['population'], ax = ax4, color ='yellow')
sns.kdeplot(minmax_df['language'], ax = ax4, color ='orange')
```

```python
from sklearn.manifold import TSNE
```

```python
tsne_robust_model = TSNE(n_components=2, random_state=0, perplexity=30.0, angle=0, n_iter=1000).fit_transform(robust_df)
tsne_robust_model.shape
```

```python
plt.figure(figsize=(12,10))
plt.title('robust t-SNE plot', fontsize=24)
sns.scatterplot(x=tsne_robust_model[:, 0], y=tsne_robust_model[:, 1], hue=target, palette='rainbow')
```

```python
tsne_standard_model = TSNE(n_components=2, random_state=0, perplexity=30.0, angle=0, n_iter=1000).fit_transform(standard_df)
tsne_standard_model.shape
```

```python
plt.figure(figsize=(12,6))
plt.title('standard t-SNE plot', fontsize=24)
sns.scatterplot(x=tsne_standard_model[:, 0], y=tsne_standard_model[:, 1], hue=target, palette='rainbow')
```

```python
tsne_minmax_model = TSNE(n_components=2, random_state=0, perplexity=30.0, angle=0, n_iter=1000).fit_transform(standard_df)
tsne_minmax_model.shape
```

```python
plt.figure(figsize=(12,6))
plt.title('minmax t-SNE plot', fontsize=24)
sns.scatterplot(x=tsne_minmax_model[:, 0], y=tsne_minmax_model[:, 1], hue=target, palette='rainbow')
```
